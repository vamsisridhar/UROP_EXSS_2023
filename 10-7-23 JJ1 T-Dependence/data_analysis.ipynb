### Automating Plotting

```python
def plot_profile(temp, ax1, ax2):

    file_path = f"{DATA_DIR}/JJ1_TDep_{temp}mK00001.dat"
    if temp == 30:
        file_path = f"{DATA_DIR}/JJ1_TDep_{temp}mK_200001.dat"

    df = pd.read_csv(file_path, skiprows=32, delimiter="\t", names= PD_NAMES)
    ax1.plot(df["Out"], df["In1"], label = "In1")
    ax1.plot(df["Out"], df["In2"], label = "In2")
    ax1.plot(df["Out"], df["In3"], label = "In3")
    ax1.plot(df["Out"], df["In4"], label = "In4")
    ax1.plot(df["Out"], df["In5"], label = "In5ax")
    ax1.plot(df["Out"], df["In1 [bwd]"], label = "In1 [bwd]")
    ax1.plot(df["Out"], df["In2 [bwd]"], label = "In2 [bwd]")
    ax1.plot(df["Out"], df["In3 [bwd]"], label = "In3 [bwd]")
    ax1.plot(df["Out"], df["In4 [bwd]"], label = "In4 [bwd]")
    ax1.plot(df["Out"], df["In5 [bwd]"], label = "In5 [bwd]ax")
    ax1.set_title(f"JJ1 {temp}mK sweep")
    ax1.set_ylabel("In (V)")
    ax1.set_xlabel("Out (V)")

    ax2.plot(df["Out"], df["LI D1"], label = "LI D1")
    ax2.plot(df["Out"], df["LI D2"], label = "LI D2")
    ax2.plot(df["Out"], df["LI D3"], label = "LI D3")
    ax2.plot(df["Out"], df["LI D4"], label = "LI D4")
    ax2.plot(df["Out"], df["LI D5"], label = "LI D5")

    ax2.plot(df["Out"], df["LI D1 [bwd]"], label = "LI D1 [bwd]")
    ax2.plot(df["Out"], df["LI D2 [bwd]"], label = "LI D2 [bwd]")
    ax2.plot(df["Out"], df["LI D3 [bwd]"], label = "LI D3 [bwd]")
    ax2.plot(df["Out"], df["LI D4 [bwd]"], label = "LI D4 [bwd]")
    ax2.plot(df["Out"], df["LI D5 [bwd]"], label = "LI D5 [bwd]")

    ax2.set_title(f"JJ1 {temp}mK sweep")
    ax2.set_ylabel("LI D3 (V)")
    ax2.set_xlabel("Out (V)")
    return (ax1, ax2)
```

```python
fig1, ax1 = plt.subplots(nrows=4, ncols=4)
fig2, ax2 = plt.subplots(nrows=4, ncols=4)
fig1.set_figheight(14)
fig1.set_figwidth(14)
fig2.set_figheight(14)
fig2.set_figwidth(14)
fig1.tight_layout(pad=3)
fig2.tight_layout(pad=3)
T_counter = 0
for i, row in enumerate(ax1):
    for j, col in enumerate(row):
        plot_profile(TEMP_VALUES[T_counter], ax1[i, j], ax2[i, j])
        T_counter += 1
plt.show()
```
