```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA_DIR = "10-7-23 JJ1 T-Dependence" 
# Out : Output 3 Setp. (V)
# In{n}: Input 1 [0000{n}] (V)
# LI D{n}: LI Demod 3 X [0000{n}] (V)
PD_NAMES = ["Out", "In1", "LI D1", "In2", "LI D2","In3", "LI D3","In4", "LI D4", "In5", "LI D5", "In1 [bwd]", "LI D1 [bwd]", "In2 [bwd]", "LI D2 [bwd]","In3 [bwd]", "LI D3 [bwd]","In4 [bwd]", "LI D4 [bwd]", "In5 [bwd]", "LI D5 [bwd]"]
```

### Exploring the 300mK run

Testing out "JJ1_TDep_300mK0001.dat"

```python
df = pd.read_csv(f"{DATA_DIR}/JJ1_TDep_60mK00001.dat", skiprows=32, delimiter="\t", names= PD_NAMES)
df.columns
```

```python
plt.plot(df["Out"], df["In1"], label = "In1")
plt.plot(df["Out"], df["In2"], label = "In2")
plt.plot(df["Out"], df["In3"], label = "In3")
plt.plot(df["Out"], df["In4"], label = "In4")
plt.plot(df["Out"], df["In5"], label = "In5")

plt.plot(df["Out"], df["In1 [bwd]"], label = "In1 [bwd]")
plt.plot(df["Out"], df["In2 [bwd]"], label = "In2 [bwd]")
plt.plot(df["Out"], df["In3 [bwd]"], label = "In3 [bwd]")
plt.plot(df["Out"], df["In4 [bwd]"], label = "In4 [bwd]")
plt.plot(df["Out"], df["In5 [bwd]"], label = "In5 [bwd]")

plt.title("JJ1 60mK sweep")
plt.ylabel("In (V)")
plt.xlabel("Out (V)")
plt.legend()
plt.show()
```
